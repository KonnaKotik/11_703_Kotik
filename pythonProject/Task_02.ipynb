```python
import numpy as np
import matplotlib.pyplot as plt
import pygame as pg
```

```python
series = np.empty((0,2))
color = (255,255,255)
pg.init()

with open("data\sunspot.txt", "r") as file:    
    
    data = file.readlines()
    for i, line in enumerate(data):
        if i > 3 and i < 3080:
            words = line.split(" ")
#             print(float(words[-1][:-1]))
            if len(words) == 1: continue
            series = np.append(series, [[i-3,float(words[-1][:-1])]], axis=0)
            
print(series)

X = series[:,0]
Y = series[:,1]
```

```python
plt.figure(figsize=(20,10))
plt.plot(X,Y)
plt.scatter(X,Y)
plt.xlim(200,2000)
plt.show()
```

```python
def moving_average(series, windowsize):
    result = []
    for n in range(1, len(series) - windowsize):
        result.append(sum(series[n+i] for i in range(windowsize))/float(windowsize))
    return result


Y_av = moving_average(Y, 20)

plt.figure(figsize=(20,10))
plt.plot(X,Y)
plt.scatter(X,Y)
plt.plot(X[:len(Y_av)], Y_av, c='r', linewidth=3)
plt.xlim(200,2000)
plt.show()

    
```
